```python
import pandas as pd

#csv data file
csv_file_path1 = 'import/2018_Chase_Loans.csv'
csv_file_path2 = 'import/2019_Chase_Loans.csv'
csv_file_path3 = 'import/2020_Chase_Loans.csv'
csv_file_path4 = 'import/2021_Chase_Loans.csv'
csv_file_path5 = 'import/2022_Chase_Loans.csv'

#read into dataframe
df = pd.read_csv(csv_file_path5)

#display
df
```

```python
#drop columns that we do not need/want or contain zero data
#shockingly large amount of the columns were not used or contained the same value regardless of outcome

to_be_dropped = ['lei', 
'preapproval', 'purchaser_type', 'rate_spread', 'denial_reason_1', 'denial_reason_2', 'denial_reason_3', 'denial_reason_4',
'total_points_and_fees', 'prepayment_penalty_term', 'debt_to_income_ratio', 'combined_loan_to_value_ratio',
'lien_status', 'applicant_credit_scoring_model', 'co_applicant_credit_scoring_model', 'hoepa_status',
'applicant_ethnicity_1','applicant_ethnicity_2','applicant_ethnicity_3','applicant_ethnicity_4','applicant_ethnicity_5',
'co_applicant_ethnicity_1','co_applicant_ethnicity_2','co_applicant_ethnicity_3','co_applicant_ethnicity_4',
'co_applicant_ethnicity_5','applicant_ethnicity_observed','co_applicant_ethnicity_observed','applicant_race_1',
'applicant_race_2','applicant_race_3','applicant_race_4','applicant_race_5','co_applicant_race_1',
'co_applicant_race_2','co_applicant_race_3','co_applicant_race_4','co_applicant_race_5','applicant_race_observed',
'co_applicant_race_observed', 'applicant_sex','co_applicant_sex','applicant_sex_observed','co_applicant_sex_observed',
'applicant_age','applicant_age_above_62','co_applicant_age','co_applicant_age_above_62', 'submission_of_application',
'balloon_payment', 'interest_only_payment', 'negative_amortization', 'other_non_amortizing_features', 'initially_payable_to_institution',
'multifamily_affordable_units', 'aus_1', 'aus_2', 'aus_3', 'aus_4', 'aus_5', 'reverse_mortgage', 'open_end_line_of_credit',
'manufactured_home_secured_property_type', 'manufactured_home_land_property_interest','census_tract',
'origination_charges', 'discount_points', 'lender_credits','intro_rate_period']

df = df.drop(columns=to_be_dropped)

df
```

```python
#converts string to numeric
df['income'] = pd.to_numeric(df['income'])

# Multiply the 'income' values by 1000 so its more human readable
df['income'] *= 1000
```

```python
#only keep rows with loans approved or denied
df = df[df['action_taken'].isin([2,3,5,6,7,8])]

#only keep actual home loans
df = df[df['loan_purpose'].isin([1])]

#fill blanks with zeros
df = df.fillna(0)

df
```

```python
#maps out our loan_type to match the key values
mapping = {
    1: 'Conventional', 
    2: 'FHA Insured', 
    3: 'VA Guaranteed', 
    4: 'Farm USDA'
}

#replace  map values
df['loan_type'] = df['loan_type'].replace(mapping)

#limits to Conventional loans
df = df[df['loan_type'].isin(['Conventional'])]
```

```python
#maps out our loan_purpose to the key values
loan_purpose_mapping = {
    1: 'Home Purchase',
    2: 'Home Improvement',
    31: 'Refinancing',
    32: 'Cash-out refinancing',
    4: 'Other Purpose',
    5: 'Not Applicable'
}

#replace  map values
df['loan_purpose'] = df['loan_purpose'].replace(loan_purpose_mapping)

df
```

```python
#maps out our construction_method to the key values
mapping = {
    1: 'Site-Built', 
    2: 'Manufactured-Home'
}

#replace  map values
df['construction_method'] = df['construction_method'].replace(mapping)

df
```

```python
#maps out our occupancy_type to the key values
mapping = {
    1: 'Principal-residence', 
    2: 'Second-residence', 
    3: 'Investment-property'
}

#replace  map values
df['occupancy_type'] = df['occupancy_type'].replace(mapping)

df
```

```python
#maps out our business_or_commercial_purpose to the key values
mapping = {
    1: 'Business', 
    2: 'Residence'
}

#replace  map values
df['business_or_commercial_purpose'] = df['business_or_commercial_purpose'].replace(mapping)

#limits our data to only Residences
df = df[df['business_or_commercial_purpose'].isin(['Residence'])]
```

```python
#maps out our action_taken to the key values
mapping = {
    1: 'Denied',  #removed
    2: 'Approved',  #approved
    3: 'Denied',  #denied
    4: 'Denied',  #removed
    5: 'Denied',  #denied
    6: 'Approved',  #approved
    7: 'Denied',  #denied
    8: 'Approved'   #approved
}
#replace  map values
df['action_taken'] = df['action_taken'].replace(mapping)

df
```

```python
#uses the FCC County Code TXT file to merge the values to the as the county_code as a new column county_name
county_mapping_df = pd.read_csv('county_codes.csv')

df = df.merge(county_mapping_df, on='county_code', how='left')

df = df.rename(columns={'county_name': 'county_name'})

#drops the now useless county_code
county_drop = ['county_code']

df = df.drop(columns=county_drop)

df
```

```python
#re-orders our data in a human reading friendly order.
data_order = [
    'activity_year',
    'loan_type',
    'action_taken',
    'loan_purpose',
    'business_or_commercial_purpose',
    'construction_method',
    'occupancy_type',
    'state_code',
    'county_name',
    'loan_amount',
    'income',
    'loan_term',
    'property_value',
    'total_units']

df = df[data_order]

df
```

```python
#Drops data with nothing so its not causing issues with the machine learning steps
df = df.dropna()

df
```
